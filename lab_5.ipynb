# Toluca Data Target->lotSize Predictors->workHours

```python
import pandas as pd

df = pd.read_csv('toluca.csv')
df
```

## Plot the predictor against the target

```python
import matplotlib.pyplot as plt
%matplotlib inline

df.plot(kind='scatter', x='workHours', y='lotSize')
```

```python
import seaborn as sns

sns.lmplot(x='workHours', y='lotSize', data=df)
```

## Plot the residual against the target

```python
r = sns.residplot('workHours', 'lotSize', data=df)
r.set(xlabel='Lot Size', ylabel='Residual')
r
```

## Determine the coefficients (slope, intercept) against the target

```python
from sklearn.linear_model import  LinearRegression

lr = LinearRegression()
lr.fit(df.drop('lotSize', axis=1), df.iloc[:,0])
lr
# df.iloc[:,1]
```

```python
print("coeficient: ", lr.coef_)
```

```python
print("intercept: ", lr.intercept_)
```

based on the plot, predictor is suitable for SLR

# Credit Data Set

```python
df = pd.read_csv('Credit.csv')
df
```

## Plot the predictor against the target

```python
df.plot(kind='scatter', x='Income', y='Limit')
```

```python
df.plot(kind='scatter', x='Rating', y='Limit')
```

```python
df.plot(kind='scatter', x='Cards', y='Limit')
```

```python
df.plot(kind='scatter', x='Age', y='Limit')
```

```python
df.plot(kind='scatter', x='Education', y='Limit')
```

## Plot the residual against the target

```python
r = sns.residplot('Income', 'Limit', data=df)
r.set(xlabel='Lot Size', ylabel='Residual')
```

```python
r = sns.residplot('Rating', 'Limit', data=df)
r.set(xlabel='Lot Size', ylabel='Residual')
```

```python
r = sns.residplot('Cards', 'Limit', data=df)
r.set(xlabel='Lot Size', ylabel='Residual')
```

```python
r = sns.residplot('Age', 'Limit', data=df)
r.set(xlabel='Lot Size', ylabel='Residual')
```

```python
r = sns.residplot('Education', 'Limit', data=df)
r.set(xlabel='Lot Size', ylabel='Residual')
```

## Determine the coefficients (slope, intercept) against the target

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression().fit(df.filter(items=['Income']),df.filter(items=['Limit']))
lin_reg
print("Income vs Limit slope: ", lin_reg.coef_)
print("Income vs Limit intercept: ", lin_reg.intercept_)
```

```python

lin_reg = LinearRegression().fit(df.filter(items=['Rating']),df.filter(items=['Limit']))
lin_reg
print("Rating vs Limit slope: ", lin_reg.coef_)
print("Rating vs Limit intercept: ", lin_reg.intercept_)
```

```python

lin_reg = LinearRegression().fit(df.filter(items=['Cards']),df.filter(items=['Limit']))
lin_reg
print("Cards vs Limit slope: ", lin_reg.coef_)
print("Cards vs Limit intercept: ", lin_reg.intercept_)
```

```python

lin_reg = LinearRegression().fit(df.filter(items=['Age']),df.filter(items=['Limit']))
lin_reg
print("Age vs Limit slope: ", lin_reg.coef_)
print("Age vs Limit intercept: ", lin_reg.intercept_)
```

```python

lin_reg = LinearRegression().fit(df.filter(items=['Education']),df.filter(items=['Limit']))
lin_reg
print("Education vs Limit slope: ", lin_reg.coef_)
print("Education vs Limit intercept: ", lin_reg.intercept_)
```

based on the plot:

1. Income and Rating are suitable for SLR.
2. Cards, Age and Education are not.
